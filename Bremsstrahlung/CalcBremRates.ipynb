```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# file = "/Users/mistryk2/Packages/nexus/Next100_Port1a_tl208_100kevents2_slim.h5"
# file = "/Users/mistryk2/Packages/nexus/Next100_Port1a_tl208_100kevents_10bar_slim.h5"
file = "/Users/mistryk2/Packages/nexus/New_tl208_slim.h5"

config = pd.read_hdf(file, "MC/configuration")
display(config)
parts  = pd.read_hdf(file, "MC/particles")
display(parts)
hits = pd.read_hdf(file, "MC/hits")
display(hits)
```

```python
gammas = parts[ (parts.particle_name == "gamma") & (round(parts.kin_energy) == round(2.6145043))]
display(gammas)

plt.figure()
plt.scatter(gammas.final_x, gammas.final_y)


print(hits.label.unique())
```

```python
# Sum the energy of the hits and bin into a histogram
hit_tot_energy = hits.groupby('event_id')['energy'].sum().reset_index(name='hit_tot_energy')
print("Num Events Tot: ", len(hit_tot_energy))
print("Num Events E>150keV: ", len(hit_tot_energy[hit_tot_energy.hit_tot_energy > 150e-3]))

# Get the hit tables and plot the total energy deposited in the TPC
uncontained_hit_events = hits[hits.label != "ACTIVE"].event_id.unique()
hits_contained_df = hits[~hits.event_id.isin(uncontained_hit_events) ]

hits_contained_df = hits

# Sum the energy of the hits and bin into a histogram
hit_tot_energy = hits_contained_df.groupby('event_id')['energy'].sum().reset_index(name='hit_tot_energy')
print("Num Events: ", len(hit_tot_energy))

# Also merge this event energy into the main dataframe
# Merge the original DataFrame with the summed energy DataFrame based on 'event'
hits_contained_df = pd.merge(hits_contained_df, hit_tot_energy, on='event_id')



plt.figure()
plt.hist(hit_tot_energy.hit_tot_energy, bins = np.arange(0,2.7,0.05), log=True);
```

```python
plt.figure()
# plt.hist(hits_contained_df.hit_tot_energy, bins = np.arange(2.53,2.7,0.05))

PP_hits = hits_contained_df[(hits_contained_df.hit_tot_energy > 2.53) & (hits_contained_df.hit_tot_energy < 2.7)  ]
PP_event_ids = PP_hits.event_id

print(len(PP_event_ids.unique()))

display(hit_tot_energy[hit_tot_energy.event_id.isin(PP_event_ids)])

plt.figure()
plt.hist(hit_tot_energy[hit_tot_energy.event_id.isin(PP_event_ids)].hit_tot_energy);
```

```python
# Get the DE peak events
DE_hits = hits_contained_df[(hits_contained_df.hit_tot_energy > 1.58) & (hits_contained_df.hit_tot_energy < 1.61)  ]
DE_event_ids = DE_hits.event_id

print(len(DE_event_ids.unique()))

plt.figure()
display(hit_tot_energy[hit_tot_energy.event_id.isin(DE_event_ids)])


plt.hist(hit_tot_energy[hit_tot_energy.event_id.isin(DE_event_ids)].hit_tot_energy);
```
