```python
# Load in the files
parts = pd.read_hdf("../../ATPC/config/ATPC_gamma.h5", 'MC/particles')

# Get events where the mother ID was a pair conversion
pair_evids = parts[ (parts.final_proc == "conv") &  (parts.primary == 1)].event_id
parts = parts[parts.event_id.isin(pair_evids)]

display(parts)

# electrons = parts[ (parts.mother_id == 1) & ((parts.particle_name == "e-") | (parts.particle_name == "e+"))]
electrons =  parts[ (parts.creator_proc == "conv") & (parts.particle_name == "e-")]
positrons =  parts[ (parts.creator_proc == "conv") & (parts.particle_name == "e+")]
```

```python
plt.figure()

plt.hist(electrons.kin_energy, bins = 50, histtype = "step", label = "electron");
plt.hist(positrons.kin_energy, bins = 50, histtype = "step", label = "positron");
plt.xlabel("Lepton energy [MeV]")
plt.ylabel("Entries")
plt.legend()

```
