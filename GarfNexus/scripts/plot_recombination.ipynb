```python
# Notebook to slim the production files for the LPR
import os
import sys
import tables as tb
import numpy  as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import dask.dataframe as dd
from matplotlib.cm import get_cmap
%config InlineBackend.figure_format = 'svg'

import re
cmap = get_cmap('Spectral')
```

```python
pressures = np.arange(2,11,1)

NEST_alpha_S1 = np.array([174237, 177330, 179161, 181148, 183844, 187251, 191317, 195930, 200950])
NEST_alpha_ie = np.array([267437, 268058, 264648, 259495, 253020, 245892, 238340, 230515, 222537])
r = np.array([0.004, 0.010, 0.020, 0.032, 0.049, 0.068, 0.091, 0.115, 0.140 ])*100

print(NEST_alpha_S1)

plt.figure()

plt.xlabel("Pressure [bar]")
plt.ylabel("Entries")

plt.scatter(pressures, NEST_alpha_S1, label = "NEST S1", color = "royalblue")
plt.axhline(y=222300, color='royalblue', linestyle='--', label = "NEXUS S1")
plt.scatter(pressures, NEST_alpha_ie, label = "NEST ie-", color = "crimson")
plt.axhline(y=244200, color='crimson', linestyle='--', label = "NEXUS ie-")

plt.title("5.49 MeV alpha, NEXT100 center, 5 bar, 1 kV/cm")

plt.legend()


plt.figure()

plt.xlabel("Pressure [bar]")
plt.ylabel("Entries")

plt.scatter(pressures, NEST_alpha_S1 + NEST_alpha_ie, label = "NEST S1 + ie- (~3% variation)", color = "royalblue")
plt.axhline(y=(222300 + 244200), color='crimson', linestyle='--', label = "NEXUS S1 + ie-")

plt.title("5.49 MeV alpha, NEXT100 center, 5 bar, 1 kV/cm")

plt.legend()

plt.figure()

plt.xlabel("Pressure [bar]")
plt.ylabel("Recombination Fraction [%]")

plt.scatter(pressures, r, label = "NEST", color = "royalblue")
plt.errorbar(10, 2.6,yerr = 1.3, label = "DEMO data (L. Serra et al 2015 JINST 10 P03025)", color = "crimson", fmt='o')
plt.title("5.49 MeV alpha, NEXT100 center, 5 bar, 1 kV/cm")
plt.legend()
```
