```python
# Notebook to study the gammas emmited from the 2nubb events
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

import os
import sys
import tables as tb
import numpy  as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
from matplotlib.colors import LogNorm  # Import LogNorm for logarithmic color scale
from matplotlib.cm import get_cmap
import dask.dataframe as dd
import glob

# Create a sequential colormap
cmap = get_cmap('Spectral')  # You can choose any sequential colormap here

plt.rcParams['figure.facecolor'] = 'w'
```

```python
# Now load in the stack 

# Directory containing HDF5 files
directory = '../files/4.3/'

# Pattern to match HDF5 files
pattern = os.path.join('../files/4.3/', '*.h5')

energies_4_3 = []
energies_5 = []
energies_15 = []

# Loop over HDF5 files in the directory
for filepath in glob.glob(os.path.join('../files/4.3/', '*.h5')):

    # Load HDF5 file into a pandas DataFrame
    hits = pd.read_hdf(filepath, "MC/hits_tot")

    energies_4_3.append(hits.hit_tot_energy.values)

# Loop over HDF5 files in the directory
for filepath in glob.glob(os.path.join('../files/5/', '*.h5')):

    # Load HDF5 file into a pandas DataFrame
    hits = pd.read_hdf(filepath, "MC/hits_tot")

    energies_5.append(hits.hit_tot_energy.values)

# Loop over HDF5 files in the directory
for filepath in glob.glob(os.path.join('../files/15/', '*.h5')):

    # Load HDF5 file into a pandas DataFrame
    hits = pd.read_hdf(filepath, "MC/hits_tot")

    energies_15.append(hits.hit_tot_energy.values)

energies_4_3 = np.concatenate(energies_4_3)
energies_5 = np.concatenate(energies_5)
energies_15 = np.concatenate(energies_15)
```

```python

plt.figure()
plt.hist(energies_4_3, bins = np.arange(0.7,2.6, 0.1), histtype = "step", color = "darkblue", label = "4.3 bar",linewidth=2);
plt.hist(energies_5, bins = np.arange(0.7,2.6, 0.1), histtype = "step", color = "purple", label = "5 bar",linewidth=2);
plt.hist(energies_15, bins = np.arange(0.7,2.6, 0.1), histtype = "step", color = "crimson", label = "15 bar",linewidth=2);
plt.legend(loc = "upper left")
plt.xlabel("Energy [MeV]")
plt.ylabel("Entries")
plt.title("1M 0nubb events in NEXT100 active")
```
